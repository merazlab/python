##While Loop

```
i = 0

while i < 5:
  print(i)
  i = i + 1
```

```
i = 0

while i < 10:
  print(i)
  i = i + 1 
  if i == 5:
    break 
```

```
i = 0

while i < 7: 
  i = i + 1  
  if i == 5:
    continue
  print(i) 
print("end")
```
