__getitem __ , __setitem __

```
class Car:
  def __init__(self):
    self.models = {}

  def add_model(self, mname, mdate):
    self.models[mname] = mdate

  def model_date(self, mname):
    return self.models[mname]

tesla = Car()
```

```
tesla.add_model("model1", 2017)
tesla.add_model("model2", 2018)
tesla.add_model("modelx", 2019)
```

```
tesla.model_date("modelx")
```

```
class Car:
  def __init__(self):
    self.models = {}

  def __setitem__(self, mname, mdate):  #(self, key, values)
    self.models[mname] = mdate
  
  def __getitem__(self, mname):
    return self.models[mname]

tesla = Car()
```

```
tesla["model1"] = 2017
tesla["model2"] = 2018
tesla["modelx"] = 2019
```

```
print(tesla["model1"])
```
