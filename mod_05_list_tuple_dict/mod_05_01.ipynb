#Basic List/Tuple/Dict



```
l = []
t = ()
d = {}

print(type(l), type(t), type(d))
```

Print value

```
l = [2, 3, 4.5,'md', "meraz"]
t = (2, 3, 4.5,'md', "meraz")
d = {"name":"meraz", "age":27, "sex":"male"}

print(l)
print(t)
print(d)
```

Access Value

```
print(l[1])
print(t[3])
print(d["age"])
```

Updating value

```
l[1] = 10
#t[3] = 8    #tupple is immutable data type, cant change
d["age"] = 30

print(l[1])
print(t[3])
print(d["age"])
```

Delete element

```
del l[1]
#del t[3]     #tupple is immutable data type, cant change
del d["age"] 

print(l)
print(t)
print(d)
```

Order

```
l = [2, 3, 1, 5]
t = (4, 1, 3 ,6)
d = {"b": 3, "c":5, "a": 1}
s = {2, 6, 1, 3, 7} #order not preserved

print(l)
print(t)
print(d)
print(s)
```

Duplicate

```
l = [2, 3, 1, 5, 2]
t = (4, 1, 3, 6, 4)
d = {"b": 3, "c":5, "a": 1, "b": 1}
s = {2, 6, 1, 3, 2}

print(l)
print(t)
print(d)
print(s)
```

For loop 

```
l = [2, 3, 1, 5]
t = (4, 1, 3 ,6)
d = {"b": 3, "c":5, "a": 1}
s = {2, 6, 1, 3, 7} #order not preserved

for i in l:
  print(i, end=" ")
print("")
for j in t:
  print(j, end=" ")
print("")

for k in d:
  print(k, end=" ")
print("")

for n in s:
  print(n, end=" ")
```

Slicing

```
l = [2, 3, 1, 5, 7, 4]
t = (4, 1, 3 ,6, 12, 10)
d = {"b": 3, "c":5, "a": 1}
s = {2, 6, 1, 3, 7} #order not preserved
print(l[-1])
print(t[-1])
#print(d[-1])
#print(s[-1])
```

```
print(l[:])
print(t[:])
```

```
print(l[1:3])
print(t[2:3])
```

```
l = [1, 2, 3, 4, 5, 6, 7, 8, 9]
t = (1, 2, 3, 4, 5, 6, 7, 8, 9)
print(l[1:4:1])
print(t[1:4:2])
```

```
l = [1, 2, 3, 4, 5, 6, 7, 8, 9]
t = (1, 2, 3, 4, 5, 6, 7, 8, 9)
print(l[-1:-5:-1])
print(t[-1:1:-1]) 
```

Length

```
l = [2, 3, 4.5,'md', "meraz"]
t = (5, 3, 3.5,'md')
d = {"name":"meraz", "age":27, "sex":"male"}

print(len(l))
print(len(t))
print(len(d))
```
