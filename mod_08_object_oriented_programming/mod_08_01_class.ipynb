```python

class MyFeatures:
    
    def __init__(self, features=None) -> None:
        self.features={}
        
    def set_features(self, name, feature):
        self.features[name] = feature
        
    def show_features(self, name):
        print(self.features[name])
        
        
        
f1 = MyFeatures()
f1.set_features('age', [2, 4, 6])
f1.set_features('saliry', [1, 2, 3])

f1.show_features('age')
f1.show_features('saliry')
```

###Construtor
-is used to define & Intialize,of non static variabel at the time of creating object.

-Constructor name is always start with __init__(self)

-Constructor is executed automatically, when object is ceated.

```python
class Demo:
  def __init__(self):
    self.a = 100
    self.b = 500
  def display(self):
    print("var a=", self.a)
    print("var b=", self.b)

  def modify(self):
    self.a = self.a + 100
  
d1 = Demo()
d1.display()
```

if you want to acces the property of one class to a another classs directly by only same class members

```python
class X: #Super or Base class
  a = 100
  def __init__(self):
    self.b = 200
  def m1(Self):
    print("m1 method in x")

class Y(X): #Derived class or Sub class
  c = 300
  def __init__(self):
    self.d = 400
  def m2(self):
    print("m2 method in x")
  def display(self):
    print(Y.c)
    print(self.d)
    self.m2()

    print(Y.a)
    print(Y.b)
    self.m1()


d1 = Y()
d1.display()
```

##Overriding

```python
class X:
  def m1(Self):
    print("in m1in x")
  def m2(self):
    print("in m2 in x")

class Y(X):
  def m1(Self):
    print("in m1 in Y")
  def m3(self):
    print("in m2 in Y")

y1 = Y()
y1.m1()
y1.m2()
y1.m3()
```

Now override the user defined class

```python
class X:
  def m1(self):
    print("m1 in x")
  
x1 = X()
print(x1)
p = x1.__str__()
print(p)
```

```python
class X:
  def m1(self):
    print("m1 in x")
  def __str__(self):
    return ("meraz")
x1 = X()
print(x1)
p = x1.__str__()
print(p)
```

##Abstraction
