For Serilization(From python to Json)

dumps() --->it serializes python dict object to json string

dump() ---> python dict object to json and write to json file

```
import json
stu ={"name": "meraz",
      "age": 35,
      "marks": 75.07,
      "ismaried": True,       #---true
      "isgf": None            #---null1
      }

jason_string = json.dumps(stu)
print(jason_string)
print(type(jason_string))  
```

```
with open("student.json", 'w') as f:
  json.dump(stu, f)
```

For Deseralization(from json to python form)

loads()--->conv json string to python dict

load()--->json to python file

```
import json
stu ='''{"name": "meraz",
      "age": 35,
      "marks": 75.07,
      "ismaried": true,      
      "isgf": null            
      }'''

stu_dict = json.loads(stu)
print(stu_dict)
print(type(stu_dict))  
```

```
for k, v in stu_dict.items():
  print("{} : {}".format(k, v))
```
