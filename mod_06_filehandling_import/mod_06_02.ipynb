File Mode

|Mode | operation    | pointer location       |
|---- | ------------------------------------ | ------------|
|r    | reading mode                          | brown **fox**  |
|w    | writing,if not exist create new file  | natural |
|x   | Open a new File                         | transparent |
|a     |open for appending at the end           |
|t     |Open in text mode                       |
|b    | open in binary mode

```
x = open("myfile.txt",)
print(x.tell()) #return curent position
print(x.read())
print(x.tell())
```

```
x = open('myfile.txt',)
x.seek(12) #seewk current position
print(x.read())
x.close()
```

```
#read data line by line
#every line a one string and store all string in list
x = open("myfile.txt")
data = x.readlines()
print(data)
x.close()
```

```
x = open("myfile.txt")
data = x.readlines()
for p in data:
  print(p)

x.close()
```

```
x = open("myfile.txt")
data = x.readlines()
for p in data:
  print(p, end="")

x.close()
```

Pickling: convert python object to a file suppported format, and object store in the file
also called seralization

```
import pickle
data =[["meraz", 789], ["dilip", 5674]]

x = open('pickle.txt', "wb") #write mode and store in binary format
pickle.dump(data, x)
x.close()
```

Unpickling

```
import pickle
x = open("pickle.txt", "rb")
data = pickle.load(x)
x.close()

print(data)
```
