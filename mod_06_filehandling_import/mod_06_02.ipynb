Pickling: convert python object to a file suppported format, and object store in the file
also called seralization

```
import pickle
data =[["meraz", 789], ["dilip", 5674]]

x = open('pickle.txt', "wb") #write mode and store in binary format
pickle.dump(data, x)
x.close()
```

Unpickling

```
import pickle
x = open("pickle.txt", "rb")
data = pickle.load(x)
x.close()

print(data)
```

#Comprehension technique
Single line statement

```
my_list = [0, 1, 2, 3, 4, 5, 6, 7]
```

```
print(my_list)
```

```
new_list = [var for var in my_list]
print(new_list)
```

```
new_list2 = [var for var in my_list if var%2 == 0]
print(new_list2)
```

New Example

```
users  = [{"name": "meraz", "age":27}, {"name": "Dilip", "age":31}, {"name": "Ritesh", "age":35}]
print(users)
```

```
name = [user["name"] for user in users]
print(name)
```

```
name2 = [user["name"] for user in users if user["age"] > 30]
print(name2)
```
