Default exception block

Handle any type off exception

Single try with multiple except block

```
try:
  x = int(input("enter first no"))
  y = int(input("enter second no"))

  z = x / y
  print("result", z)
except(ZeroDivisionError):
  print("second value is not zero")
except(ValueError):
  print("Value eroor")
except:
  print("error occured")

print("end")
```

Raise exception

-Default exception block is last block always

```
try:
  x = int(input("enter first no"))
  y = int(input("enter second no"))

  z = x / y
  print("result", z)
  raise KeyError
except(ZeroDivisionError):
  print("second value is not zero")
except(ValueError):
  print("Value eroor")
except:
  print("error occured")

print("end")
```

##finally

The sets of statements which are compulsary to exectute whether exception is raised or not raised

Exception- Only user frienly statement

Resource releasing statemnet are written in finally group

```
try:
  x = int(input("enter first no"))
  y = int(input("enter second no"))

  z = x / y
  print("result", z)
except(ZeroDivisionError):
  print("second value is not zero")
finally:  #run exception raised or not raised
  print("bye")

print("end")
```
